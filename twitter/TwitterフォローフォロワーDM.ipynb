```python
def thousand_split(df):
    # /1000 で1000区切りで何分割できるか検証
    split_num = math.ceil(len(df)/1000)
    
    split_list = []
    
    # 分割開始
    for i in range(split_num):
        
        # 1000区切りできるものは分割
        if i < split_num:
            split_cell = df[i*1000: (i+1)*1000].reset_index(drop=True)
        
        #1000区切りでいない最後のものはコチラで分割
        elif i == split_num:
            split_cell = df[i*1000:].reset_index(drop=True)
            
        split_list.append(split_cell)
        
    return split_list
```

```python
def change_shape(df):
    #初期設定
    box1 = []
    box1 =  np.array(box1)
    box1 = box1.reshape(-1,)
    box2 = []
    box2 =  np.array(box2)
    box2 = box2.reshape(-1,)
    
    check_len = len(df)
    index_num = int(np.trunc(check_len /100))
    
    #100で割り切れる数じゃなかったら分割する
    if int(check_len%100) !=0:
        df1 = df.loc[:index_num*100-1]
        df2 = df.loc[index_num*100:]
        print(df1.shape)
        
        #shapeの変更
        for i in range(index_num):
            dff1 = np.array(df1[i*100: i*100+100])
            box1 = np.hstack((box1,dff1))
        
        #shapeの変更
        for j in range(1):
            dff2 = np.array(df2[0:])
            box2 = np.hstack((box2,dff2))
        
        box1 = box1.reshape(-1,100)
        box2 = box2.reshape(1,-1)
        
        print(box1.shape)
        
        box1 = pd.DataFrame(box1.T)
        box2 = pd.DataFrame(box2.T)
        
        return box1, box2
            
    elif int(check_len%100) == 0:
        
        df1 =  df.loc[:index_num*100-1]
        #shapeの変更
        for i in range(index_num):
            dff = np.array(df1[i*100: i*100+100])
            box1 = np.hstack((box1,dff))
            
        box1 = box1.reshape(-1,100)
        box2 = box2.reshape(-1,100)
        
        box1 = pd.DataFrame(box1.T)
        
        return box1, box2
        
            
    else:
        print('dfのshapeが(-1,)の形状になっているか確認して下さい')
        
        return box1, box2
```

```python
#ログイン処理時に認証が入る可能性があるので
def  login_check(account_id:list, account_pass:list):
    twitter = GetTwitter()
    for i in range(len(account_id)):
        print("{}番目のaccount".format(i+1))
        driver = twitter.login(account_id[i], account_pass[i])
        # ここで認証を解除する
        time.sleep(20)
        print("20秒停止します")
        driver.quit()
    print("finish")
```

### 一気に処理を回す  

1000区切りでない情報の部分が処理できないエラーが発生したので893など1000未満の数字の部分のみ取り出してmainで回すことを推奨

```python
def main_loop(boxs1:list, box2, account_id:list, account_pass:list):
    li_fin = []
    
    for i in range(len(boxs1)+1):
        try:
            if i < len(boxs1):
                li, li1, li2, li3, li4 = main(account_id, account_pass,0 , boxs1[i])
                lili = [li, li1, li2, li3, li4]
            elif i == len(boxs1):
                li, li1, li2, li3, li4 = main(account_id, account_pass, 0, box2)
                lili = [li, li1, li2, li3, li4]
            li_fin.append(lili)
            
        except KeyboardInterrupt:
            print("処理を中断しました")
            lili  = ["処理中断{}".format(i)]
            li_fin.append(lili)
        except:
            lili = [i]
            li_fin.append(lili)
            
    return li_fin        
```
